```python
new=df[df.reviews >=10000]
len(new)
```

```python
new=new[new.reviews <=60000]
len(new)
```

```python
rev=[]
for i in new['reviews']:
    rev.append(i)
    
score=[]
for i in new['score']:
    score.append(i)    
```

```python
plt.figure(figsize=(20,10))
```
