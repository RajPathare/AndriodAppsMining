```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df = pd.read_csv('dataset/education.csv',index_col=0)
```

```python
# Groupby by country
reviews = df.groupby("reviews")

# Summary statistic of all countries
reviews.describe().head()
```

```python
reviews.mean().sort_values(by="score",ascending=False).head()
```

```python
plt.figure(figsize=(15,10))
reviews.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Wines")
plt.show()
```

```python
# Start with one review:
text = df.description[0]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("first_review_wordcloud.png")
```

```python
low=df[df.score<=3.5]
high=df[df.score>=4.5]
```

```python
text = " ".join(review for review in low.description)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(8,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
w=[]
n=[]
def freq(str): 
  
    # break the string into list of words 
    str_list = str.split() 
  
    # gives set of unique words 
    unique_words = set(str_list) 
      
    for words in unique_words : 
        w.append(words)
        n.append(str_list.count(words))

freq(text)        
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=50)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Wines")
plt.plot(w,n)
```
